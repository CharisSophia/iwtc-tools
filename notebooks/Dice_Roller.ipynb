```python
from pathlib import Path
import sys

here = Path.cwd()
print("Notebook CWD:", here)

# Try to locate lib/dice.py from *this* folder
dice_here = here / "lib" / "dice.py"
dice_parent = here.parent / "lib" / "dice.py"

print("Exists here/lib/dice.py?    ", dice_here.exists())
print("Exists parent/lib/dice.py?  ", dice_parent.exists())

# Choose which path to add:
if dice_here.exists():
    sys.path.insert(0, str(here))           # lib is alongside the notebook
    print("Added to sys.path:", here)
elif dice_parent.exists():
    sys.path.insert(0, str(here.parent))     # notebook is likely in notebooks/, lib is one level up
    print("Added to sys.path:", here.parent)
else:
    raise FileNotFoundError("Could not find lib/dice.py relative to this notebook.")
```
